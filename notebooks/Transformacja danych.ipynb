```python
import pandas as pd
import os
from datetime import datetime
```

```python
path = os.path.join("..", "data", "interim", "job_offers_2024_07_07.csv")
df = pd.read_csv(path, sep=";")


df.head()
```

Dodawanie Å›redniej

```python
df['average_salary'] = df[['salary_low', 'salary_high']].mean(axis=1)
```

```python
df.head()
```

Zapis do katalogu

```python
path = os.path.join("..", "data", "processed")

os.makedirs(path, exist_ok=True)

current_date = datetime.today().strftime('%Y_%m_%d')
filename = os.path.join(path, f"job_offers_{current_date}.csv")
df.to_csv(filename, sep=';', encoding='utf-8', index=False)

print(f"Saved all job offers to: {filename}")
```

```python
filename = os.path.join(path, f"job_offers_{current_date}.json")

df.to_json(filename, orient='records', lines=True, force_ascii=False)

print(f"Saved all job offers to: {filename}")
```
